```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Set seed for reproducibility
np.random.seed(0)

# Generate class 0 data (centered at (2, 2))
x0 = np.random.randn(50, 2) + np.array([2, 2])
y0 = np.zeros(50) - 1

# Generate class 1 data (centered at (6, 6))
x1 = np.random.randn(50, 2) + np.array([6, 6])
y1 = np.ones(50)

# Combine the data
X = np.vstack((x0, x1))
y = np.concatenate((y0, y1))

# Create a DataFrame
df = pd.DataFrame(X, columns=['feature1', 'feature2'])
df['label'] = y.astype(int)

# Plot data points
plt.scatter(df[df.label == -1]['feature1'], df[df.label == -1]['feature2'], label='Class 0')
plt.scatter(df[df.label == 1]['feature1'], df[df.label == 1]['feature2'], label='Class 1')

# Plot decision boundary
x_vals = np.linspace(X[:, 0].min()-1, X[:, 0].max()+1, 100)
# decision boundary: w0*x + w1*y + b = 0  solve for y

W=[0.6917911820312496, 0.9495201139235399];   b=-6.239518290575923
y_vals = -(W[0] * x_vals + b) / W[1]

plt.plot(x_vals, y_vals, 'k--', label='Decision Boundary')

# Formatting
plt.legend()
plt.xlabel('feature1')
plt.ylabel('feature2')
plt.title('Logistic Regression Decision Boundary')
plt.grid(True)
plt.show()
```
